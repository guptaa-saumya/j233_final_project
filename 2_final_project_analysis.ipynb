### 5. Graphs for analysis

Now that all the tables are ready, I went ahead and made several types of charts.

```python
allfires = total_fires.reset_index().melt('Years')
allfires = allfires.drop(allfires[allfires['variable'] == 'index'].index)
alt.Chart(allfires, title = "Number of Fires in Residential Homes").mark_line().encode(
    x='Years',
    y='value',
    color='variable')
```

```python
firecost = fire_cost.reset_index().melt('Years')
firecost = firecost.drop(firecost[firecost['variable'] == 'index'].index)
alt.Chart(firecost, title = "Fire Dollar Loss for Residential Homes").mark_line().encode(
    x='Years',
    y='value',
    color='variable')
```

### 6. My analysis

Like I mentioned in my cleaning up notebook, I wanted to see what was causing fires the most for people. So when I figured out which ones it was, I narrowed down my scope and wanted to see what the effects of those fires are. 

Based off my first graph, I wanted to see the trend of fire rates throughout the years. Was it consistent? Did things change and if they did why did they change? Most of the data was pretty much consistent over a period of time except for there being some decreases in heating fires for both multifamily homes and 1 & 2 family homes. According to the definition of the causes, this probably means that children are playing with heat sources less or people in general or not playing with heat sources or fire that much anymore. 

Because the amount of fires was pretty much consistent year after year, I expected to see the same for the fire dollar loss. However the data didn't really follow those lines. The most costly cause of fire was an electrical malfunction, even though that didn't cause nearly as many fires as cooking and playing with a heat source. However, this probably isn't suprising because there seems to be some research out there that electrical fires tend to cost the most out of other fires in terms of money, life, and losing valuables (https://www.afcisafety.org/files/realcostelectricalfire.pdf. 

Besides those graphs, my main focus was on the other ones, which dives into the causes of the fires and the injuries and deaths that occur from them. I did five different charts to show the five different causes I focused on. Based on these graphs, cooking fires seem to cause more injuries than death, which is probably explained by the fact that people are usually attentive and awake when they or someone in the building is cooking which means that people can evacuate faster. The amount of deaths did increase a little bit for fires caused by playing with a heat source. This can probably be explained by how children are usually the ones to be killed in these fires, maybe because they are the ones playing with the heat source or for some othe reason (https://sfm.illinois.gov/content/dam/soi/en/web/sfm/sfmdocuments/documents/homevictimsfactsheet.pdf). The last graph was a little bit suprising though. Even though there were less fires caused by electrical malfunction than cooking, it caused more injuries and deaths. This probably goes hand in hand with costs. Sometimes people aren't prepared for these types of fires or are not using equipment in their house that can increase their chance of escaping, which can explain the higher numbers of death and injury (https://www.afcisafety.org/files/realcostelectricalfire.pdf). Overall, I think exploring the electrical malfunctions peaked my interest and can be something more for me to look into.
